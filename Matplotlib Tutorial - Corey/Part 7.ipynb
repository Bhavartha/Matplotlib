### 1) Basic

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.style.use('seaborn-pastel')

plt.scatter(x, y, s=sizes, c=colors, edgecolors='black',
            lw=1, alpha=0.7, cmap='Greens')
c_bar = plt.colorbar()
c_bar.set_label('Satisfaction')
plt.show()
```

### 2) Using Data from CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (16, 8)
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('data (Part 7).csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

data.head()
```

```python
plt.scatter(view_count, likes, lw=1, alpha=0.75,
            edgecolors='black', c=ratio, cmap='viridis')
plt.xscale('log')
plt.yscale('log')
c_bar = plt.colorbar(pad=0.05)
c_bar.set_label('Likes-Dislike Ratio',labelpad=20)

plt.show()
```
