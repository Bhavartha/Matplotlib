### Basic Plot ###

**Parameters**

c: color

lw: linewidth

ls: linestyle

marker
<hr>

We can also use format-string like `<Marker><Color><Linestyle>` (interchangable)

```python
plt.plot(ages_x, py_dev_y, label='Python Devs', c='blue', lw=3)
plt.plot(ages_x, js_dev_y, label='JS Devs', c='#adad3b', lw=3)
plt.plot(ages_x, dev_y, label='All Devs', c='#444444', ls='--')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.grid()
plt.show()
```
