# Filling Area on Line Plots
<hr>

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (16, 8)
```

```python
data = pd.read_csv('data (Part 5).csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

data.head()
```

**Basic Plot**

```python
plt.style.use('fivethirtyeight')

plt.plot(ages, py_salaries, label='Python Dev Salary')
plt.plot(ages, dev_salaries, ls='--', label='Dev Salary')
plt.xlabel('Ages', labelpad=20)
plt.ylabel('Salary', labelpad=20)
plt.title('Comparison of Salaries')
plt.legend()
plt.show()
```

<hr>

**Advanced Plot: Python Dev vs Dev Salaries**

*Uses fill to denote when the python devs salary is more and when the salary is less*

```python
data[['Python', 'All_Devs']].head(5)
```

```python
plt.style.use('fivethirtyeight')

plt.plot(ages, py_salaries, label='Python Dev Salary')
plt.plot(ages, dev_salaries, ls='--', label='Dev Salary')

plt.fill_between(ages, py_salaries, dev_salaries, alpha=0.35,
                 where=(py_salaries > dev_salaries), interpolate=True,label='Above avg')
plt.fill_between(ages, py_salaries, dev_salaries, alpha=0.5,
                 where=(py_salaries <= dev_salaries), interpolate=True,label='Below avg')

plt.xlabel('Ages', labelpad=20)
plt.ylabel('Salary', labelpad=20)
plt.title('Comparison of Salaries')

plt.legend()
plt.show()
```
