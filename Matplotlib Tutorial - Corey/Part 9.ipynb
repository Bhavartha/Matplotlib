# Plotting Live Data in Real-Time
<hr>

```python
%matplotlib notebook

from itertools import count
from random import randint
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
plt.style.use('fivethirtyeight')

index = count()
x_vals=[]
y_vals=[]

def animate(i,x,y):
    x_vals.append(next(index))
    y_vals.append(randint(0,5))
    
    #To avoid compression of data
    x= x_vals[-10:] 
    y = y_vals[-10:]
    
    plt.cla() #Clear axis
    plt.plot(x,y)

ani= FuncAnimation(plt.gcf(),animate,interval=1000,fargs=(x_vals,y_vals))

plt.show()
```
