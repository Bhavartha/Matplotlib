## Matplotlib example workflow

```python
# 0. importing 
import matplotlib.pyplot as plt
%matplotlib inline

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Simple plot',xlabel='x-axis',ylabel='y-axis')

# 5. Save and show
# fig.savefig('fig.png')
```

```python
# Making plot from dict

nut_butter_prices = {"Almond butter":10,
                    "Peanut butter":8,
                    "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Butter prices",ylabel="Price ($)");
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Butter prices",xlabel="Price ($)");
```

```python
# Make some data for histograms

x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
x = np.random.randn(1000)
d1 = np.random.randint(0,10,20)
d2 = np.random.randint(0,5,20) 
```

```python
# Option 1

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2)
ax2.hist(x)
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.scatter(d1,d2);
```

```python
# Option 2

fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Plot to each different index

ax[0,0].plot(x,x/2)
ax[0,1].hist(x)
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].scatter(d1,d2);
```

```python
# On different dataset
import pandas as pd 

heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
# Histogram of age

heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots=True,figsize=(16,32));
```

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50
```

```python
#Pyplot method
over_50.plot.scatter(x='age',y='chol',c='target',figsize=(16,8));
```

```python
# OO method (jugaad)

fig,ax = plt.subplots()
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax,figsize=(16,8));
ax.set_xlim([45,80]);
```

```python
# OO method

fig,ax = plt.subplots(figsize=(16,8))

#plot data
scatter = ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax.set(title='Heart disease and cholestrol levels', xlabel='Age',ylabel='Cholestrol')
ax.axhline(over_50['chol'].mean(),ls='--')

ax.legend(*scatter.legend_elements(),title='Target');
```

```python
over_50.head()
```

```python
# Subplot of chol,age,thalach

fig,(ax0,ax1) = plt.subplots(figsize=(16,10),nrows=2,ncols=1,sharex=True)

# Add data to first figure
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax0.set(title='Heart disease and cholestrol levels',ylabel='Cholestrol')
ax0.legend(*scatter.legend_elements(),title='Target')
ax0.axhline(over_50['chol'].mean(),ls='--');

# Add data to second figure
scatter = ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax1.set(title='Heart disease and cholestrol levels', xlabel='Age',ylabel='Thalach')
ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(over_50['thalach'].mean(),ls='--');
```
